```python
import matplotlib.pyplot as plt
```

```python
pdbID = '6cne'
text_file = open(pdbID+".pdb."+str(5)+"."+str(4)+".ThermoDescript", 'r')
constants = text_file.readlines()
text_file.close()
```

```python
pCOREX = []
for c in constants:
    pCOREX.append(float(c.split()[1]))
```

```python
pdbID = '6cne'
text_file = open(pdbID + ".pdb." + str(5) + "." + str(4) + ".ThermoDescriptMC" ,'r')
constants = text_file.readlines()
text_file.close()
```

```python
COREX = []
for c in constants[:]:
    COREX.append(float(c.split()[1]))
```

```python
text_file = open('6cne.pdb.W5.T25S0.500.ThermoDescript.Original', 'r')
constants = text_file.readlines()
text_file.close()

oCOREX = []
for c in constants[1:]:
    oCOREX.append(float(c.split()[-3]))
```

```python
x = range(1, len(COREX) + 1)
```

```python
plt.plot(x, oCOREX, '--o', label = "COREX (Hilser et al.))(9208 micro-states)")
plt.plot(x, pCOREX, '--o', label = "pCOREX (9208 micro-states)")
plt.plot(x,COREX, '--o', color = '#ff66cc', label = "pCOREXMC(4000 samples)")
plt.ylim(0, max(max(COREX), max(pCOREX)) + 2)
plt.xlabel("Residue Number")
plt.ylabel("Stability Constant $(ln k_{f})$")
plt.legend()
plt.title("6cne.pdb")
plt.savefig("6cneMC.pdf")
plt.show()
```
